# **Sort the data by product and sub-product**

```python
# Sort the data

# Sorting by product and subproduct
# From filtering the dataset on the CFPB website, I already knew there were 4 product categories
# This code was created with the help of Google Gemini, which is built into Google Colab
grouped_products = df.groupby('Product')['Sub-product'].agg(['nunique', lambda x: x.unique().tolist()])
grouped_products.columns = ['Number of Sub-products', 'List of Sub-products']

# Reset index to make 'Product' a column and add 'Product number'
grouped_products = grouped_products.reset_index()
grouped_products.insert(0, 'Product number', range(1, 1 + len(grouped_products)))

display(grouped_products)
```

```python
# Seeing how many consumer complaints there are for each product and sub-product
# This code was created with the help of Google Gemini, which is built into Google Colab
complaints_per_product_subproduct = df.groupby(['Product', 'Sub-product']).size().reset_index(name='Number of Complaints')
display(complaints_per_product_subproduct)
```

# **Draw conclusions**

From these results, I decided to prioritise the dataset and consumer complaints by picking a single subcategory for each product with the highest number of complaints.

My thought process here was to try to find a way to ensure that the identified topics are as accurate and reliable as possible.

This is because in phase two I identified the most prevalent topics across the corpus but as one can see from the table above this would most likely not be an accurate reflection of the data and the consumer comlaints.

# **Sort the proritised data by issue**

```python
# Here I used Gemini to help me turn each of the prioritised data groups above into its own dataset.

# Filter for Credit card (General-purpose credit card or charge card)
credit_card_df = df[(df['Product'] == 'Credit card') &
                    (df['Sub-product'] == 'General-purpose credit card or charge card')].copy()

# Filter for Debt collection (I do not know)
debt_collection_df = df[(df['Product'] == 'Debt collection') &
                        (df['Sub-product'] == 'I do not know')].copy()

# Filter for Mortgage (Conventional home mortgage)
mortgage_df = df[(df['Product'] == 'Mortgage') &
                 (df['Sub-product'] == 'Conventional home mortgage')].copy()

# Filter for Student loan (Federal student loan servicing)
student_loan_df = df[(df['Product'] == 'Student loan') &
                     (df['Sub-product'] == 'Federal student loan servicing')].copy()

print(f"Credit card (General-purpose credit card or charge card) complaints: {len(credit_card_df)}")
display(credit_card_df.head())

print(f"Debt collection (I do not know) complaints: {len(debt_collection_df)}")
display(debt_collection_df.head())

print(f"Mortgage (Conventional home mortgage) complaints: {len(mortgage_df)}")
display(mortgage_df.head())

print(f"Student loan (Federal student loan servicing) complaints: {len(student_loan_df)}")
display(student_loan_df.head())
```

```python
# Sort the data
# Here I wanted to explore the issues and sub-issues of the prioritised data
# Here I once again used Google Gemini

# Issues and Sub-issues for Credit Card (General-purpose credit card or charge card)
print("\n--- Issues and Sub-issues for Credit Card (General-purpose credit card or charge card) ---")
grouped_credit_card_issues = credit_card_df.groupby('Issue')['Sub-issue'].agg(['nunique', lambda x: x.unique().tolist()])
grouped_credit_card_issues.columns = ['Number of Sub-issues', 'List of Sub-issues']
print(f"Total Number of Issues: {len(grouped_credit_card_issues)}")
display(grouped_credit_card_issues.reset_index())
```

```python
print("\n--- Number of Complaints per Issue for Credit Card (General-purpose credit card or charge card) ---")
credit_card_issues_complaint_count = credit_card_df.groupby('Issue').size().reset_index(name='Number of Complaints')
display(credit_card_issues_complaint_count.sort_values(by='Number of Complaints', ascending=False))
```

```python
# Issues and Sub-issues for Debt Collection (I do not know)
print("\n--- Issues and Sub-issues for Debt Collection (I do not know) ---")
grouped_debt_collection_issues = debt_collection_df.groupby('Issue')['Sub-issue'].agg(['nunique', lambda x: x.unique().tolist()])
grouped_debt_collection_issues.columns = ['Number of Sub-issues', 'List of Sub-issues']
print(f"Total Number of Issues: {len(grouped_debt_collection_issues)}")
display(grouped_debt_collection_issues.reset_index())
```

```python
print("\n--- Number of Complaints per Issue for Debt Collection (I do not know) ---")
debt_collection_issues_complaint_count = debt_collection_df.groupby('Issue').size().reset_index(name='Number of Complaints')
display(debt_collection_issues_complaint_count.sort_values(by='Number of Complaints', ascending=False))
```

```python
# Issues and Sub-issues for Mortgage (Conventional home mortgage)
print("\n--- Issues and Sub-issues for Mortgage (Conventional home mortgage) ---")
grouped_mortgage_issues = mortgage_df.groupby('Issue')['Sub-issue'].agg(['nunique', lambda x: x.unique().tolist()])
grouped_mortgage_issues.columns = ['Number of Sub-issues', 'List of Sub-issues']
print(f"Total Number of Issues: {len(grouped_mortgage_issues)}")
display(grouped_mortgage_issues.reset_index())
```

```python
print("\n--- Number of Complaints per Issue for Mortgage (Conventional home mortgage) ---")
mortgage_issues_complaint_count = mortgage_df.groupby('Issue').size().reset_index(name='Number of Complaints')
display(mortgage_issues_complaint_count.sort_values(by='Number of Complaints', ascending=False))
```

```python
# Issues and Sub-issues for Student loan (Federal student loan servicing)
print("\n--- Issues and Sub-issues for Student loan (Federal student loan servicing) ---")
grouped_student_loan_issues = student_loan_df.groupby('Issue')['Sub-issue'].agg(['nunique', lambda x: x.unique().tolist()])
grouped_student_loan_issues.columns = ['Number of Sub-issues', 'List of Sub-issues']
print(f"Total Number of Issues: {len(grouped_student_loan_issues)}")
display(grouped_student_loan_issues.reset_index())
```

```python
print("\n--- Number of Complaints per Issue for Student loan (Federal student loan servicing) ---")
student_loan_issues_complaint_count = student_loan_df.groupby('Issue').size().reset_index(name='Number of Complaints')
display(student_loan_issues_complaint_count.sort_values(by='Number of Complaints', ascending=False))
```
